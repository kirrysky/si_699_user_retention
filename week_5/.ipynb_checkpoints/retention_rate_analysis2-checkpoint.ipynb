```python
daily={'user_id':[]}
for i in range(1,85):
    daily['day'+str(i)]=[]
```

```python
for j in range(len(user_everyday_info_new2)):
    if j%100000==0:
        print('still...working...no..worries')
    for i in range(len(user_everyday_info_new2.iloc[j])):
        if i==0:
            daily['user_id'].append(user_everyday_info_new2.iloc[j][i])
            k=1
        else:
            if i%5==1:
                daily['day'+str(k)].append(user_everyday_info_new2.iloc[j][i])
                k+=1
```

```python
daily_form=pd.DataFrame.from_dict(daily)
```
