```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

```python
order_=pd.read_csv('OrderAprtoJun.csv')
```

```python
order_.head()
```

```python
order_.info()
```

```python
order_['create_time']=pd.to_datetime(order_['create_time'])
```

```python
final_day=pd.to_datetime("2016-6-13 00:00:00")
```

```python
# 截取Final day之前的日子
order_2=order_[(final_day-order_['create_time']).dt.days>=0]
```

```python
order_2[pd.isnull(order_2['send_time'])]
```

```python
# 第一次place order距离predict time的天数
```

```python
first_order=order_2[['user_id','create_time']].groupby(["user_id"]).min().reset_index()
first_order.head(15)
```

```python
first_order['first_order']=(final_day-first_order['create_time']).dt.days
```

```python
first_order2=first_order[['user_id','first_order']]
first_order2.head()
```

```python
# 最后一次place order距离predict time的天数
last_order=order_2[['user_id','create_time']].groupby(["user_id"]).max().reset_index()
last_order.head(15)
```

```python
last_order['last_order']=(final_day-last_order['create_time']).dt.days
last_order2=last_order[['user_id','last_order']]
last_order2.head()
```

```python
# distance min,max,avg
```

```python
distance=order_2[['user_id','distance']]
distance_mean=distance.groupby(["user_id"]).mean().reset_index()
distance_mean.columns=['user_id','distance_mean']
distance_mean.head()
```

```python
distance_max=distance.groupby(["user_id"]).max().reset_index()
distance_max.columns=['user_id','distance_max']
distance_max.head()
```

```python
distance_min=distance.groupby(["user_id"]).min().reset_index()
distance_min.columns=['user_id','distance_min']
distance_min.head()
```

```python
# send_time min max avg
```

```python
send_time=order_2[['user_id','send_time','create_time','update_time']]
send_time.head(15)
```

```python
send_time['send_time']=pd.to_datetime(send_time['send_time'])
send_time['create_time']=pd.to_datetime(send_time['create_time'])
send_time['update_time']=pd.to_datetime(send_time['update_time'])
```

```python
send_time['overall_time']=send_time['send_time']-send_time['create_time']
send_time['update2create_time']=send_time['update_time']-send_time['create_time']
send_time['send2update_time']=send_time['send_time']-send_time['update_time']
```

```python
send_time.head()
```

```python
send_time['overall_time']=send_time['overall_time'].dt.total_seconds()
send_time['update2create_time']=send_time['update2create_time'].dt.total_seconds()
send_time['send2update_time']=send_time['send2update_time'].dt.total_seconds()
```

```python
send_time.head()
```

```python
send_time2=send_time[['user_id','overall_time','update2create_time','send2update_time']]
```

```python
send_time2.head()
```

```python
send_time2_mean=send_time2.groupby(["user_id"]).mean().reset_index()
send_time2_mean.columns=['user_id','overall_time_mean','update2create_time_mean','send2update_time_mean']
send_time2_mean.head()
```

```python
send_time2_max=send_time2.groupby(["user_id"]).max().reset_index()
send_time2_max.columns=['user_id','overall_time_max','update2create_time_max','send2update_time_max']
send_time2_max.head()
```

```python
send_time2_min=send_time2.groupby(["user_id"]).min().reset_index()
send_time2_min.columns=['user_id','overall_time_min','update2create_time_min','send2update_time_min']
send_time2_min.head()
```
