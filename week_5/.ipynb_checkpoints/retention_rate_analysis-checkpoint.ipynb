```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

```python
training_data=pd.read_csv('training_data.csv')
```

```python
training_data.head()
```

```python
len(training_data)
```

```python
labels=pd.read_csv('daily_form4.csv')
```

```python
labels.head()
```

```python
labels['user_id']=labels['user_id'].astype(int)
```

```python
labels['label']=labels['day6']>0
```

```python
labels['label']=labels['label'].astype(int)
```

```python
labels.head()
```

```python
labels.columns=['user_id','2w_sum1','2w_sum2','2w_sum3','2w_sum4','2w_sum5','2w_sum6','label']
```

```python
labels=labels.drop(['2w_sum6'],axis=1)
```

```python
labeled_data=pd.merge(training_data,labels,how='inner',on='user_id')
```

```python
labeled_data.head()
```

```python
#shuffle
shuffled_labeled_data = labeled_data.sample(frac=1).reset_index(drop=True)
```

```python
#shuffle twice
shuffled_labeled_data = shuffled_labeled_data.sample(frac=1).reset_index(drop=True)
```

```python
shuffled_labeled_data.head()
```

```python
import matplotlib.pyplot as plt
nan_value = shuffled_labeled_data.isnull().sum()
nan_value = nan_value[nan_value > 0]
nan_value.sort_values(inplace=True,ascending=False)
nan_value.plot.bar()
```

```python
shuffled_labeled_data['user_invite_code']=shuffled_labeled_data['user_invite_code'].fillna('-1')
```

```python
shuffled_labeled_data['user_occupation']=shuffled_labeled_data['user_occupation'].fillna('Missing')
```

```python
shuffled_labeled_data['user_age']=shuffled_labeled_data['user_age'].astype(int)
```

```python
shuffled_labeled_data['user_age']=shuffled_labeled_data['user_age'].fillna('Missing')
```

```python
shuffled_labeled_data['send2update_time_min']=shuffled_labeled_data['send2update_time_min'].fillna('-1')
shuffled_labeled_data['update2create_time_min']=shuffled_labeled_data['update2create_time_min'].fillna('-1')
shuffled_labeled_data['send2update_time_max']=shuffled_labeled_data['send2update_time_max'].fillna('-1')
shuffled_labeled_data['update2create_time_max']=shuffled_labeled_data['update2create_time_max'].fillna('-1')
shuffled_labeled_data['send2update_time_mean']=shuffled_labeled_data['send2update_time_mean'].fillna('-1')
shuffled_labeled_data['update2create_time_mean']=shuffled_labeled_data['update2create_time_mean'].fillna('-1')
```

```python
shuffled_labeled_data['kitchen_age']=shuffled_labeled_data['kitchen_age'].fillna('Missing')
```

```python
from sklearn.model_selection import train_test_split
```

```python
shuffled_labeled_data.to_csv('shuffled_labeled_data.csv',index=False)
```
