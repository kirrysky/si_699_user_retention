```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

```python
order_=pd.read_csv('../OrderAprtoJun.csv')
```

```python
order_.head()
```

```python
order_['create_time']=pd.to_datetime(order_['create_time'])
```

```python
final_day=pd.to_datetime("2016-6-13 00:00:00")
```

```python
# 截取Final day之前的日子
order_2=order_[(final_day-order_['create_time']).dt.days>=0]
```

```python
# 最常去的kitchen下的总订单数
user_kitchen=order_2.groupby(["user_id","kitchen_id"]).size().reset_index()
user_kitchen.columns=['user_id','kitchen_id','order_nums']
user_kitchen.head(15)
```

```python
user_kitchen_unique=user_kitchen.groupby(["user_id"]).size().reset_index()
user_kitchen_unique.columns=['user_id','unique_kitchen_order_nums']
user_kitchen_unique.head(15)
```

```python
# kitchen_royalty
```

```python
last_user=user_kitchen.iloc[0]['user_id']
```

```python
users={'user_id':[],'total_order_nums':[]}
```

```python
max_order_nums=0
```

```python
for i in range(1,len(user_kitchen)):
    user=user_kitchen.iloc[i]['user_id']
    order_nums=user_kitchen.iloc[i]['order_nums']
    if user!=user_kitchen.iloc[i-1]['user_id']:
        users['user_id'].append(user_kitchen.iloc[i-1]['user_id'])
        users['total_order_nums'].append(max_order_nums)
        max_order_nums=order_nums
    else:     
        max_order_nums+=order_nums
```

```python
user
```

```python
user_kitchen[user_kitchen['user_id']==user]
```

```python
# 把最后一个加进去
users['user_id'].append(user)
# users['kitchen_id'].append(kitchen)
users['total_order_nums'].append(order_nums)
```

```python
total_kitchen=pd.DataFrame.from_dict(users)
```

```python
total_kitchen.head(15)
```

```python
total_kitchen['kitchen_royalty']=total_kitchen['total_order_nums']/user_kitchen_unique['unique_kitchen_order_nums']
```

```python
total_kitchen[['user_id','kitchen_royalty']].to_csv('kitchen_royalty.csv',index=False)
```

```python
most_freq_kitchen=pd.DataFrame.from_dict(users)
```
