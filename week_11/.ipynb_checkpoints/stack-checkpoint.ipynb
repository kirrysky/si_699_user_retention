```python
nn_test=pd.read_csv('nn_test.csv')
```

```python
nn_test.columns=["nn_test"]
```

```python
lr_pred=pd.read_csv('logistic_regression1.csv')
```

```python
lr_pred.columns=["lr_pred"]
```

```python
rf_pred=pd.read_csv('random_forest.csv')
```

```python
rf_pred.columns=["rf_pred"]
```

```python
true_label=pd.read_csv('true_label.csv')
```

```python
true_label
```

```python
# all_=[lightgbm_pred,nn_test]
all_=[lightgbm_pred,nn_test,rf_pred]
```

```python
res=pd.concat(all_,axis=1)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(res,true_label,test_size=0.2,random_state=42)
```

```python
y_trian_pred = gnb.fit(X_train, y_train).predict_proba(X_train)
```

```python
y_test_pred = gnb.fit(X_train, y_train).predict_proba(X_test)
```

```python
roc_auc_score(y_train,y_trian_pred.T[1])
```

```python
roc_auc_score(y_test,y_test_pred.T[1])
```

```python
lightgbm_=np.where(lightgbm_pred["lightgbm_pred"]>=0.5,1,0)
```

```python
ridge_=np.where(ridge_pred["ridge_pred"]>=0.5,1,0)
```

```python
nn_=np.where(nn_test["nn_test"]>=0.5,1,0)
```

```python
lr_=np.where(lr_pred["lr_pred"]>=0.5,1,0)
```

```python
rf_=np.where(rf_pred["rf_pred"]>=0.5,1,0)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(true_label, lightgbm_))
```

```python
print(metrics.classification_report(true_label, ridge_))
```

```python
print(metrics.classification_report(true_label, nn_))
```

```python
print(metrics.classification_report(true_label, lr_))
```

```python
print(metrics.classification_report(true_label, rf_))
```

```python
counts_=[]
for i in range(len(rf_)):
    count=0
    if rf_[i]==1:
        count+=1
    if lr_[i]==1:
        count+=1
    if nn_[i]==1:
        count+=1
    if ridge_[i]==1:
        count+=1
    if lightgbm_[i]==1:
        count+=1
    counts_.append(count)
```

```python
voting=np.where(np.array(counts_)>=4,1,0)
```

```python
print(metrics.classification_report(true_label, voting))
```

```python
pd.DataFrame(voting).to_csv('voting.csv',index=False)
```
